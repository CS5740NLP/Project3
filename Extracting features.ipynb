```python
import csv
new_data = []
with open('data.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        new_data.append(row)
```

```python
contexts = [i[0] for i in new_data]
questions = [i[1] for i in new_data]
candidate_sentences = [i[2] for i in new_data]
answers = [i[3] for i in new_data]
labels = [i[4] for i in new_data]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer  
vec = TfidfVectorizer()
voc = vec.vocabulary_
a = vec.fit_transform(contexts)
```

```python
# Matching Word Frequencies
MWF = []
for i,question in enumerate(questions):
    sum = 0
    for word in question.split():
        if word in candidate_sentences[i].split():
            if word.lower() in voc:
                index = voc[word.lower()]
                sum += a[i,index]
    MWF.append(sum)
```

```python
# Span Word Frequencies
SWF = []
for i,sentence in enumerate(candidate_sentences):
    sum = 0
    for word in sentence.split():
        if word.lower() in voc:
            index = voc[word.lower()]
            sum += a[i,index]
    SWF.append(sum)
```

```python
# Lengths
count=0
length = []
for i,sentence in enumerate(candidate_sentences):
    if answers[i] in sentence:
        index = sentence.index(answers[i])
    else:
        index = len(sentence)//2
    left = len(sentence[:index].split())
    inside = len(answers[i].split())
    whole = len(sentence.split())
    length.append((left,inside,whole))
```

```python
def decode(dev_data):
    dev_contexts_index = []
    dev_questions = []
    dev_map = {}
    for i,line in enumerate(dev_data):
        dev_contexts_index.append(line[0])
        dev_questions.append(line[1])
        dev_candidates = []
        for j in range(2,len(line),3):
            dev_candidates.append((line[j],line[j+1],line[j+2]))
        dev_map[i]=dev_candidates
    return dev_contexts_index,dev_questions,dev_map
```
